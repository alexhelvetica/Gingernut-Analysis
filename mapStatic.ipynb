```python
import json
import pandas
import matplotlib.pyplot
import geopandas
import contextily

skuSetName = "gingernut"
storeResults = json.load(open(f"{skuSetName}Coles5.json"))
skus = json.load(open(f"{skuSetName}Sku.json"))

storesDataFrame = pandas.DataFrame.from_dict(storeResults.values())

storesGeoDataFrame = geopandas.GeoDataFrame(
    storesDataFrame, geometry=geopandas.points_from_xy(storesDataFrame.longitude, storesDataFrame.latitude),
     crs="WGS84"
)

fig, ax = matplotlib.pyplot.subplots(figsize=(15,15))

color,marker = 0, 4
for sku in skus.values():
    storesGeoDataFrame[
    storesGeoDataFrame[str(sku["id"])] != "Not Stocked"
        ].plot(ax=ax, 
                markersize=20, 
                color=f"C{color}", 
                marker=marker, 
                label=f"{sku['display-name']}")
    marker +=1
    color +=1

ax.legend(prop={'size':15})
contextily.add_basemap(ax, crs="WGS84")
```
